```python
import gym
import random
```

```python
env = gym.make("FrozenLake-v0") # Acrobat-v1, MountainCar-v0, FrozenLake-v0
```

```python
for i in range(20):
    env.reset() # Reset basically starts a new episode
    for t in range(100): # 100 time intervals for each episode
        
        # Render visualization to screen
        env.render()
        
        # Execute some action from sample action space; go left or right
        env.step(env.action_space.sample())
```

```python
env.close()
```
